### Procedure for making  Tool : 1
We search information in Tavily and Extract 5 relevant links from the search results.

```python

tavily_client = TavilyClient(api_key=os.getenv("TAVILY_API_KEY"))

@tool
def web_search(query : str) -> str:
    """Search the web for recent and reliable information on a topic . Returns Titles , URLs and snippets."""
    results = tavily_client.search(query=query,max_results=5)

    out = []

    for r in results['results']:
        out.append(
            f"Title: {r['title']}\nURL: {r['url']}\nSnippet: {r['content'][:300]}\n"
        )
    
    return "\n----\n".join(out)

```

```python
# Checking  Above tool 
tavily_result = web_search.invoke("Jantar mantar protest")
print(tavily_result)
print(tavily_result.results)
```

### Procedure for making  Tool : 2

As the tavily give little bit content based on  the link, we will use link and the beautiful soup to extract the content of the link 

I use AI to make scrapper as i dont know to use beautiful soup. Not Bad :) 

```python
@tool
def scrape_url(url: str) -> str:
    """Scrape and return clean text content from a given URL for deeper reading."""
    try:
        resp = requests.get(url, timeout=8, headers={"User-Agent": "Mozilla/5.0"})
        soup = BeautifulSoup(resp.text, "html.parser")
        for tag in soup(["script", "style", "nav", "footer"]):
            tag.decompose()
        return soup.get_text(separator=" ", strip=True)[:3000]
    except Exception as e:
        return f"Could not scrape URL: {str(e)}"
```

```python
# Checking Tool
scrap_result = scrape_url.invoke("https://www.bbc.com/news/articles/cly0ey5y10qo")
print(scrap_result)
```

## Tool Making Process Is Done

Now we will use the tools to make a agent which can search and extract the content of the link and give us the answer of our query.

Libraries required for making an agent

```python
from langchain.agents import create_agent
from langchain_mistralai import ChatMistralAI
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser
```

```python
#model setup 
llm = ChatMistralAI(model = "mistral-small-2506",temperature=0)
```

```python
# Making  search Agent 
def build_search_agent():
    return create_agent(
        model = llm,
        tools= [web_search]
    )
```

```python
# Making Web scrapper agent
def build_reader_agent():
    return create_agent(
        model = llm,
        tools = [scrape_url]
    )
```

### Now we create a writer 


```python
writer_prompt = ChatPromptTemplate.from_messages([
    ("system", "You are an expert research writer. Write clear, structured and insightful reports."),
    ("human", """Write a detailed research report on the topic below.

Topic: {topic}

Research Gathered:
{research}

Structure the report as:
- Introduction
- Key Findings (minimum 3 well-explained points)
- Conclusion
- Sources (list all URLs found in the research)

Be detailed, factual and professional."""),
])
```

#### Creating the writer Pipeline/chain

```python
# Writer Chain

writer_chain = writer_prompt | llm | StrOutputParser() # we use parser for formatted/beautiful output

# So when we invoke the writer_chain , we have to pass 2 parameters : - topic(topic we give ) and research(partial+deep research)

```

---

### Now we will make a Critic pipeline that will review the result generated by the writer_chain

```python
# Critic Prompt

critic_prompt = ChatPromptTemplate.from_messages([
     ("system", "You are a sharp and constructive research critic. Be honest and specific."),
    ("human", """Review the research report below and evaluate it strictly.

Report:
{report}

Respond in this exact format:

Score: X/10

Strengths:
- ...
- ...

Areas to Improve:
- ...
- ...

One line verdict:
..."""),
])

```

```python
# Critic pipeline
critic_chain = critic_prompt | llm | StrOutputParser()

# when we invoke it we provide it the input of writer_chain
```

### Now we will create a final pipeling that will call agents in proper manner and use the writer and critic to generate the final answer of the query.

NOTE: We can also use a shared storage to store the intermediate results of the agents and use them in the final pipeline.

```python
from langchain_core.messages import HumanMessage



def run_research_pipeline(topic : str) -> dict:

    shared_storage = {}

    #search agent working 
    print("\n"+" ="*50)
    print("step 1 - search agent is working ...")
    print("="*50)


    # Calling the 1 agent : That use tavily and do partial Search
    search_agent = build_search_agent()
    search_result = search_agent.invoke({
        "messages": [HumanMessage(content=f"Find recent, reliable and detailed information about: {topic}")]
    })
    shared_storage["search_results"] = search_result['messages'][-1].content

    #print("\n search result ",shared_storage['search_results'])




    # Calling the 2 agent : That use deepResearch using Beautifulsoapp scrapping
    print("\n"+" ="*50)
    print("step 2 - Reader agent is scraping top resources ...")
    print("="*50)

    reader_agent = build_reader_agent()
    reader_result = reader_agent.invoke({
    "messages": [HumanMessage(content= 
            f"Based on the following search results about '{topic}', "
            f"pick the most relevant URL and scrape it for deeper content.\n\n"
            f"Search Results:\n{shared_storage['search_results'][:800]}"
            )]
    })

    shared_storage["scrapped_content"] = reader_result['messages'][-1].content

    #print("\nscraped content: \n", shared_storage['scraped_content'])


    # Now we need a structured Output based on all the information Scrapped

    print("\n"+" ="*50)
    print("step 3 - Writer is drafting the report ...")
    print("="*50)

    research_combined = (
        f"SEARCH RESULTS : \n {shared_storage['search_results']} \n\n"
        f"DETAILED SCRAPED CONTENT : \n {shared_storage['scrapped_content']}"
    )

    writer_report = writer_chain.invoke({
        "topic":topic,
        "research":research_combined
    })

    shared_storage["report"] = writer_report

    #print("\n Final Report\n",shared_storage['report'])

   


    #critic report based on the report

    print("\n"+" ="*50)
    print("step 4 - critic is reviewing the report ")
    print("="*50)

    shared_storage["feedback"] = critic_chain.invoke({
        "report":shared_storage['report']
    })

    #print("\n critic report \n", shared_storage['feedback'])

    return shared_storage


    



    
```

```python
if __name__ == "__main__":
    topic = "Protest on jantar Mantar"
    final_result = run_research_pipeline(topic)
```

Now we have a final Report

```python
print("Partial Search\n")
print(final_result["search_results"])


print("Deep  Research\n")
print(final_result["scrapped_content"])


print("Report By Writer\n")
print(final_result["report"])


print("What Critic Says-\n")
print(final_result["feedback"])

```
